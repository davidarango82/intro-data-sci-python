```python
import numpy as np
import pandas as pd

pd.options.display.max_columns = 999
# pd.options.display.max_rows = 999
pd.set_option('display.max_colwidth', -1)

archivo = 'pptos_faltantes.csv'
#crono_df = pd.read_csv(archivo, encoding = 'unicode_escape', sep=',')
crono_df = pd.read_csv(archivo, encoding = 'UTF-8', sep=',')

##### Presupuestos que hay en el archivo #####
pptos = crono_df['Presupuesto'].unique()
```

```python
rows=crono_df.shape[0]

for row in range(rows):
    
    temp_ppto=crono_df.loc[row,'Presupuesto']

    # TEST "SEM 1": 
    if ((crono_df.loc[row,'ttmto_homol']=='PROFILAXIS')
        |(crono_df.loc[row,'ttmto_homol']=='DETARTRAJE')
        |(crono_df.loc[row,'ttmto_homol']=='FERULIZACION')
        |(crono_df.loc[row,'ttmto_homol']=='RESINA FOTOPOLIMERIZADO 1 SUP POSTERIOR Y ANTERIOR')
        |(crono_df.loc[row,'ttmto_homol']=='RESINA FOTOPOLIMERIZADO 2 SUP POSTERIOR Y POSTERIOR')
        |(crono_df.loc[row,'ttmto_homol']=='RESINA FOTOPOLIMERIZADO 3 SUP POSTERIOR O POSTERIOR')
        |(crono_df.loc[row,'ttmto_homol']=='TAC 1 Arcada')
        |(crono_df.loc[row,'ttmto_homol']=='TAC 2 Arcadas')
        |(crono_df.loc[row,'ttmto_homol']=='DESOPTURACION PARA NUCLEO')
        |(crono_df.loc[row,'ttmto_homol']=='TRATAMIENTO DE CONDUCTO UNIRADICULAR')
        |(crono_df.loc[row,'ttmto_homol']=='RETRATAMIENTO ENDODONCIA BIRADICULAR')
        |(crono_df.loc[row,'ttmto_homol']=='RETRATAMIENDO ENDODONCIA MULTIRADICULAR')
        |(crono_df.loc[row,'ttmto_homol']=='TRATAMIENTO DE CONDUCTO BIRADICULAR')
        |(crono_df.loc[row,'ttmto_homol']=='RETRATAMIENTO ENDODONCIA UNIRADICULAR')
        |(crono_df.loc[row,'ttmto_homol']=='TRATAMIENTO DE CONDUCTO MULTIRADICULAR')
        |(crono_df.loc[row,'ttmto_homol']=='REPARACION DE PT(LABORATORIO)')
        |(crono_df.loc[row,'ttmto_homol']=='CITA TALLADO SELECTIVO X PIEZA')
        |(crono_df.loc[row,'ttmto_homol']=='REPARACION ESTANDAR (REPARACION METALICOS)')
        |(crono_df.loc[row,'ttmto_homol']=='ESTUDIO CEFALOMETRICO')
        |(crono_df.loc[row,'ttmto_homol']=='EXODONCIA DE RESTOS RADICULARES')
        |(crono_df.loc[row,'ttmto_homol']=='EXODONCIA SIMPLE COMPLEJIDAD 1-2-3')
        |(crono_df.loc[row,'ttmto_homol']=='EXODONCIA RESTO RADICULAR QUIRURGICO')
        |(crono_df.loc[row,'ttmto_homol']=='VENTANA QUIRURGICA')
        |(crono_df.loc[row,'ttmto_homol']=='TORNILLO DE EXPANSION')
        |(crono_df.loc[row,'ttmto_homol']=='PLACA DE LEVANTE')
        |(crono_df.loc[row,'ttmto_homol']=='PROFILAXIS PERIODONTAL')
        |(crono_df.loc[row,'ttmto_homol']=='REPARACION TUBO ESTANDAR')
        |(crono_df.loc[row,'ttmto_homol']=='PROTESIS TEMP')
        |(crono_df.loc[row,'ttmto_homol']=='PROTESIS TEMPORAL  4/8 DIENTES')
        |(crono_df.loc[row,'ttmto_homol']=='PROTESIS TEMPORAL 1')
        |(crono_df.loc[row,'ttmto_homol']=='RETIRO BRACKETS ARCADA SUPERIOR')
        |(crono_df.loc[row,'ttmto_homol']=='RETIRO BRACKETS ARCADA INFERIOR')
        |(crono_df.loc[row,'ttmto_homol']=='FLUOR')
        |(crono_df.loc[row,'ttmto_homol']=='REPARACION AUTOLIGADO METALICO') 
        |(crono_df.loc[row,'ttmto_homol']=='FRENILLECTOMIA LINGUAL Y LABIAL')
        |(crono_df.loc[row,'ttmto_homol']=='CAMBIO DE NILON (POR UNIDAD DENTAL) SOBREDENTADURA CON BARRA(INCLUY')       
        ):
        crono_df.loc[row,'Semana'] = 1

    # TEST "SEM 2 Y TRATAMIENTOS QX":
    elif((crono_df.loc[row,'ttmto_homol']=='NUCLEOS')|(crono_df.loc[row,'ttmto_homol']=='PROVISIONALES')
        |(crono_df.loc[row,'ttmto_homol']=='INJERTO OSEO GRANULADO 05 GRMS')
        |(crono_df.loc[row,'ttmto_homol']=='MEMBRANA REABSORBIBLE')
        |(crono_df.loc[row,'ttmto_homol']=='INJERTO OSEO GRANULADO  05 GRMS')
        |(crono_df.loc[row,'ttmto_homol']=='INJERTO OSEO GRANULADO 1 GRAMO')
        |(crono_df.loc[row,'ttmto_homol']=='REGENERACION OSEA (SIN HUESO NI')
        |(crono_df.loc[row,'ttmto_homol']=='ELEVACION  AL SENO MAXILAR TRAUMATICA (INCLUYE INJERTO OSEO  + MEMBRANA)')
        |(crono_df.loc[row,'ttmto_homol']=='EXODONCIA QUIRURGICA DE CORDAL (INCLUIDO SEMIERUPCIONADO)')
        |(crono_df.loc[row,'ttmto_homol']=='EXODONCIA CORDAL SIMPLE')
        |(crono_df.loc[row,'ttmto_homol']=='BLANQUEAMIENTO DIENTE NO VITAL (TRATAMIENTO COMPLETO)')
        ):
        crono_df.loc[row,'Semana'] =2

    # TEST "RETENEDORES": 
    elif((crono_df.loc[row,'ttmto_homol']=='RETENEDOR ESSIX O ESTETICOS')
    |(crono_df.loc[row,'ttmto_homol']=='RETENEDOR FIJO 3X3 EN ALAMBRE')
    |(crono_df.loc[row,'ttmto_homol']=='RETENEDOR CONTINUO')
    ):
        if crono_df[(crono_df['Presupuesto']==temp_ppto)&(
        (crono_df['ttmto_homol']=='MENSUALIDAD ORTODONCIA ZAFIRO')
        |(crono_df['ttmto_homol']=='MENSUALIDAD ORTODONCIA METALICA')
        |(crono_df['ttmto_homol']=='MENSUALIDAD ORTODONCIA ESTETICOS O CERAMICOS')
        |(crono_df[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA AUTOLIGABLES')
        |(crono_df[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA AUTOLIGADOS')
        )].count()[0]:
            temp = crono_df[(crono_df['Presupuesto']==temp_ppto)&(
            (crono_df['ttmto_homol']=='MENSUALIDAD ORTODONCIA ZAFIRO')
            |(crono_df['ttmto_homol']=='MENSUALIDAD ORTODONCIA METALICA')
            |(crono_df['ttmto_homol']=='MENSUALIDAD ORTODONCIA ESTETICOS O CERAMICOS')
            |(crono_df[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA AUTOLIGABLES')
            |(crono_df[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA AUTOLIGADOS')
            )]
            temp.reset_index(inplace=True)
            crono_df.loc[row,'Semana']=temp.loc[0].at['Cantidad Presupuestada']+temp.loc[0].at['Semana']
        else:
            crono_df.loc[row,'Semana']=1
        
    # TEST "MENS ORTO 30 DIAS":
    elif ((crono_df.loc[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA METALICA')
    |(crono_df.loc[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA ZAFIRO')
    |(crono_df.loc[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA ESTETICOS O CERAMICOS')
    |(crono_df.loc[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA 4X2')
    ):              
        crono_df.loc[row,'Semana']=6
        
    # TEST "MENS ORTO 45 DIAS":
    elif ((crono_df.loc[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA AUTOLIGABLES')
    |(crono_df.loc[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA AUTOLIGADOS')
    ):
        crono_df.loc[row,'Semana']=8
        
    # TEST "MONTAJES DE ORTODONCIA":
    elif ((crono_df.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA SUPERIOR AUTOLIGABLES')
    |(crono_df.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA SUPERIOR ZAFIRO')
    |(crono_df.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA INFERIOR ZAFIRO')
    |(crono_df.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA INFERIOR AUTOLIGABLES')
    |(crono_df.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA INFERIOR AUTOLIGADOS')
    |(crono_df.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA SUPERIOR AUTOLIGADOS')
    |(crono_df.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA INFERIOR ESTETICOS O CERAMICOS') 
    |(crono_df.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA SUPERIOR ESTETICOS O CERAMICOS')
    |(crono_df.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA INFERIOR METALICOS')
    |(crono_df.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA SUPERIOR METALICOS')
    ):
        crono_df.loc[row,'Semana']=2        
        
    # TEST "BLANQUEAMIENTO":
    elif ((crono_df.loc[row,'ttmto_homol']=='BLANQUEAMIENTO DENTAL 1')
    |(crono_df.loc[row,'ttmto_homol']=='BLANQUEAMIENTO DENTAL 2')):
        crono_df.loc[row,'Semana']=1

    # TEST "RASPAJE CON ALISADO RADICULAR":
    elif ((crono_df.loc[row,'ttmto_homol']=='RASPAJE CON ALISADO RADICULAR A CAMPO ABIERTO X CUADRANTE')):
        crono_df.loc[row,'Semana']=1
    
    # TEST "IMPLANTE y ELEVACION AL SENO ATRAUMATICA":
    elif ((crono_df.loc[row,'ttmto_homol']=='IMPLANTE')|(crono_df.loc[row,'ttmto_homol']=='ELEVACION AL SENO ATRAUMATICA (POR CUADRANTE)')):
        if crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='REGENERACION OSEA (SIN HUESO NI')].count()[0]:
            temp = crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='REGENERACION OSEA (SIN HUESO NI')]
            temp.reset_index(inplace=True)
            crono_df.loc[row,'Semana']=temp.loc[0].at['Semana']+24
        else:
            crono_df.loc[row,'Semana']=2
    
    # TEST "REHAB SOBRE IMPLANTES":
    elif ((crono_df.loc[row,'ttmto_homol']=='SOBREDENTADURA SOBRE 3 IMPLANTES')
    |(crono_df.loc[row,'ttmto_homol']=='SOBREDENTADURA SOBRE 4 IMPLANTES')
    |(crono_df.loc[row,'ttmto_homol']=='HIBRIDA SOBRE 6 IMPLANTES')
    |(crono_df.loc[row,'ttmto_homol']=='CORONA SOBRE IMPLANTE')
    |(crono_df.loc[row,'ttmto_homol']=='CORONA SOBRE IMPLANTE ZIRCONIO')
    |(crono_df.loc[row,'ttmto_homol']=='PROVISIONAL ESTETICO ATORNILLADO CON PERFIL DE')
    ):
        if crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='IMPLANTE')].count()[0]:
            temp = crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='IMPLANTE')]
            temp.reset_index(inplace=True)
            crono_df.loc[row,'Semana']=temp.loc[0].at['Semana']+16
        else:
            crono_df.loc[row,'Semana']=1
    
    # TEST "REHAB SOBRE DIENTES"  
    elif ((crono_df.loc[row,'ttmto_homol']=='CARILLA ESTETICA EN RESINA')
    |(crono_df.loc[row,'ttmto_homol']=='CORONA METAL CERAMICA')
    |(crono_df.loc[row,'ttmto_homol']=='INCRUSTACION EN CEROMERO')
    |(crono_df.loc[row,'ttmto_homol']=='CORONA CON COLLAR')
    |(crono_df.loc[row,'ttmto_homol']=='CORONAS EN ZIRCONIO SISTEMA CEREC')
    ):
        if crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='BLANQUEAMIENTO DENTAL 1')].count()[0]:
            temp = crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='BLANQUEAMIENTO DENTAL 1')]
            temp.reset_index(inplace=True)
            crono_df.loc[row,'Semana']=temp.loc[0].at['Semana']+3
        else:
            crono_df.loc[row,'Semana']=2

    # TEST "PLACA NEUROMIORELAJANTE y PLACA DE BRUXISMO":
    elif ((crono_df.loc[row,'ttmto_homol']=='PLACA NEUROMIORELAJANTE ACTIVADA')|(crono_df.loc[row,'ttmto_homol']=='PLACA DE BRUXISMO')):
        if crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='CORONA SOBRE IMPLANTE ZIRCONIO')].count()[0]:
            temp = crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='CORONA SOBRE IMPLANTE ZIRCONIO')]
            temp.reset_index(inplace=True)
            crono_df.loc[row,'Semana']=temp.loc[0].at['Semana']+4
        if crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='CORONA SOBRE IMPLANTE')].count()[0]:
            temp = crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='CORONA SOBRE IMPLANTE')]
            temp.reset_index(inplace=True)
            crono_df.loc[row,'Semana']=temp.loc[0].at['Semana']+4
        if crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='CARILLA ESTETICA EN RESINA')].count()[0]:
            temp = crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='CARILLA ESTETICA EN RESINA')]
            temp.reset_index(inplace=True)
            crono_df.loc[row,'Semana']=temp.loc[0].at['Semana']+4
        else:
            crono_df.loc[row,'Semana']=1

    # TEST "PROTESIS REMOVIBLES":
    elif ((crono_df.loc[row,'ttmto_homol']=='PROTESIS METALICA REMOVIBLE')
    |(crono_df.loc[row,'ttmto_homol']=='PROTESIS TOTAL SUPERIOR O INFERIOR')):
        if crono_df[(crono_df['Presupuesto']==temp_ppto)
        &((crono_df['ttmto_homol']=='EXODONCIA DE RESTOS RADICULARES')
        |(crono_df['ttmto_homol']=='EXODONCIA RESTO RADICULAR QUIRURGICO'))].count()[0]:
            temp = crono_df[(crono_df['Presupuesto']==temp_ppto)&((crono_df['ttmto_homol']=='EXODONCIA DE RESTOS RADICULARES')
            |(crono_df['ttmto_homol']=='EXODONCIA RESTO RADICULAR QUIRURGICO'))]
            temp.reset_index(inplace=True)
            crono_df.loc[row,'Semana']=temp.loc[0].at['Semana']+6
        else:
            crono_df.loc[row,'Semana']=1
            
    # TEST "INJERTO GINGIVAL LIBRE":
    elif ((crono_df.loc[row,'ttmto_homol']=='INJERTO GINGIVAL LIBRE')):
        crono_df.loc[row,'Semana']=1            
        
revisiones = crono_df[crono_df['Semana'] == 0]
print('Presupuestos no-ordenados: ',revisiones['Nº Historia'].unique())
revisiones[['Nº Historia','ttmto_homol']]
#crono_df
```
