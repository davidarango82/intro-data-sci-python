```python
import numpy as np
import pandas as pd

pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
pd.set_option('display.max_colwidth', -1)

archivo_crono = 'crono_Dentix_Calle_90_1.csv'
crono_df = pd.read_csv(archivo_crono, encoding = 'unicode_escape', sep=',')

##### Poner en 1's las semanas y reordenar. ESTO ES TEMPORAL #####
crono_df['Semana'] = 1

##### Extraer los ttmtos de cada presupuesto ##### (aplicamos un FOR)

#crono_df.set_index('Presupuesto',inplace=True)
#crono_df.loc[crono_df.index.unique()[0]]
pptos = crono_df['Presupuesto'].unique()
ppto_temp = crono_df[crono_df['Presupuesto'] == pptos[0]]
#ppto_temp
crono_df
```

```python
### esto es para pruebas! ###
crono_df.reset_index(inplace=True)
crono_df.loc[0,'pieza']=1
crono_df
#############################


crono_df.set_index(['Presupuesto'],inplace=True)

rows = crono_df.loc[pptos[0]].shape[0] #calcular filas del ttmto para iteraciones

for i in range(rows):

    pieza_ttmto=crono_df.loc[pptos[0]].pieza.iloc[i]
    if (crono_df.loc[pptos[0]].ttmto_homol.iloc[i]=='DESOPTURACION PARA NUCLEO'):
        print('desob: ok',i)
        if ((crono_df.loc[pptos[0]].ttmto_homol == 'RESINA FOTOPOLIMERIZADO 1 SUP POSTERIOR Y ANTERIOR')*1).sum()>=1:
            print('resina: ok',i)
            #aquí: extraer pieza de resina
            crono_df.loc[pptos[0],'Semana'] = 3
        else: 
            print('desob: no',i)
            crono_df.loc[pptos[0],'Semana'] = 2
    else: 
        print('fin ppto',i)
        print('pieza',pieza_ttmto,)

#crono_df

#((crono_df.loc[pptos[0]].ttmto_homol == 'DESOPTURACION PARA NUCLEO')*1).sum()

```
