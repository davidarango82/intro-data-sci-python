```python
#EL ARCHIVO CARGADO DEBE ACTUALIZARSE AL MENOS MENSUALMENTE CON LOS PACIENTES NUEVOS

import numpy as np
import pandas as pd

from IPython.display import display
pd.options.display.max_columns = None

pd.set_option("display.max_rows",1000)

bday_df = pd.read_csv('ClinicCloud Ventas Con FechaNac.csv', engine='python')
bday_df.head(3)
```

```python
bday_df['bday']=None
bday_df['fecha_ref']='30/07'

bday_df['bday']=bday_df['fechanac'].str.slice(start=0,stop=5)
bday_df=bday_df[bday_df['bday']==bday_df['fecha_ref']]
#bday_df=bday_df.set_index('Nombre_Clinica')
bday_df=bday_df[['Nombre_Clinica','nombre','apellidos','tele1','tele2','fechanac','TotalPagos','Fecha Presupuesto']]
bday_df=bday_df.reset_index()
bday_df.head()
```

```python
# generacion de reportes

clinicas = bday_df['Nombre_Clinica'].unique()

for i in clinicas:
    temp_df = bday_df[bday_df['Nombre_Clinica'] == i]
    temp_df.to_csv('cumple %s.csv' % i, index = False)                  
                              
bday_df.to_csv('cumpleanos.csv', index = False)
print("Archivos generados!")
```
