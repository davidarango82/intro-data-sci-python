```python
import numpy as np
import pandas as pd
import time

archivo = 'Produccion_pendiente_exigible_con_tratamientos_Dentix_Calle_90.csv'
PPE = pd.read_csv(archivo, encoding = 'unicode_escape', sep=',')
fecha_inicio = pd.read_csv('fecha_inicio.csv', encoding = 'UTF-8', sep=',')
#fecha_inicio
```

```python
#Eliminación de Puntos y Revisiones del PPE y duplicados
PPE.set_index('Servicio', inplace=True)
PPE.drop(index='Puntos y Revisiones.  ',inplace=True)
PPE.reset_index(inplace=True)
PPE['duplicated'] = PPE.duplicated()
PPE.set_index('duplicated',inplace=True)
PPE.drop(True,inplace=True)
PPE.reset_index(inplace=True)
PPE.drop(columns=['duplicated'],inplace=True)
PPE
```

```python
# Asignación de Fecha de Inicio a la PPE (fecha de aceptación de cada ppto)
PPE=pd.merge(PPE,fecha_inicio,how='left',left_on=['Nº Historia','Presupuesto'],right_on=['Nº Historia','Presupuesto'])
PPE['fecha_inicio'].fillna(PPE['Fecha Pres.'],inplace=True)
PPE
```

```python
#PPE.reset_index(inplace=True)
PPE['Cantidad Presupuestada'] = (PPE['Producidos']+PPE['Pendientes Producción'])/PPE['Importe Pres.']
PPE['Cantidad Presupuestada'] = np.ceil(PPE['Cantidad Presupuestada'])
PPE['Cantidad Presupuestada'].fillna(1,inplace=True) #las CSI traen vacío el campo 'Pendientes Produccion'.
# en realidad parece que las unicas cantidades presupuestadas q son necesarias son las de Mensualidades de ortodoncia.
# Para CSI, el %Realizado es el que nos sirve para calcular Cantidad Producida.

PPE['ttmto_homol'] = PPE['Servicio'].str.slice(0,-13)
PPE['ttmto_homol'] = PPE['ttmto_homol'].str.strip()
PPE['pieza'] = PPE['Servicio'].str.slice(-2)
PPE['pieza'] = PPE['pieza'].str.strip()
PPE['fecha_inicio'] = time.strftime("%d/%m/%Y")
PPE['Semana']=0

#PPE = PPE[['Clínica', 'Fecha Pres._x', 'Nº Historia', 'Presupuesto', 
#                                   'Servicio', 'Semana', 'Cantidad Presupuestada','Doctor_x', 
#                                   'Explicado_x', 'ttmto_homol','pieza', 'fecha_inicio']]

print('Numero de tratamientos a ordenar:',PPE.count()[0])

PPE.set_index(['Clínica','Nº Historia','Servicio'])
```

```python
rows=PPE.shape[0]

for row in range(rows):
    
    temp_ppto=PPE.loc[row,'Presupuesto']

    # TEST "SEM 1": 
    if ((PPE.loc[row,'ttmto_homol']=='PROFILAXIS')
        |(PPE.loc[row,'ttmto_homol']=='DETARTRAJE')
        |(PPE.loc[row,'ttmto_homol']=='FERULIZACION')
        |(PPE.loc[row,'ttmto_homol']=='RESINA FOTOPOLIMERIZADO 1 SUP POSTERIOR Y ANTERIOR')
        |(PPE.loc[row,'ttmto_homol']=='RESINA FOTOPOLIMERIZADO 2 SUP POSTERIOR Y POSTERIOR')
        |(PPE.loc[row,'ttmto_homol']=='RESINA FOTOPOLIMERIZADO 3 SUP POSTERIOR O POSTERIOR')
        |(PPE.loc[row,'ttmto_homol']=='SELLANTES POR FISURAS POR DIENTE')
        |(PPE.loc[row,'ttmto_homol']=='TAC 1 Arcada')
        |(PPE.loc[row,'ttmto_homol']=='TAC 2 Arcadas')
        |(PPE.loc[row,'ttmto_homol']=='DESOPTURACION PARA NUCLEO')
        |(PPE.loc[row,'ttmto_homol']=='TRATAMIENTO DE CONDUCTO UNIRADICULAR')
        |(PPE.loc[row,'ttmto_homol']=='RETRATAMIENTO ENDODONCIA BIRADICULAR')
        |(PPE.loc[row,'ttmto_homol']=='RETRATAMIENDO ENDODONCIA MULTIRADICULAR')
        |(PPE.loc[row,'ttmto_homol']=='TRATAMIENTO DE CONDUCTO BIRADICULAR')
        |(PPE.loc[row,'ttmto_homol']=='RETRATAMIENTO ENDODONCIA UNIRADICULAR')
        |(PPE.loc[row,'ttmto_homol']=='TRATAMIENTO DE CONDUCTO MULTIRADICULAR')
        |(PPE.loc[row,'ttmto_homol']=='REPARACION DE PT(LABORATORIO)')
        |(PPE.loc[row,'ttmto_homol']=='CITA TALLADO SELECTIVO X PIEZA')
        |(PPE.loc[row,'ttmto_homol']=='REPARACION ESTANDAR (REPARACION METALICOS)')
        |(PPE.loc[row,'ttmto_homol']=='ESTUDIO CEFALOMETRICO')
        |(PPE.loc[row,'ttmto_homol']=='EXODONCIA DE RESTOS RADICULARES')
        |(PPE.loc[row,'ttmto_homol']=='EXODONCIA SIMPLE COMPLEJIDAD 1-2-3')
        |(PPE.loc[row,'ttmto_homol']=='EXODONCIA RESTO RADICULAR QUIRURGICO')
        |(PPE.loc[row,'ttmto_homol']=='EXODONCIA SIMPLE EN DIENTES TEMPORALES')
        |(PPE.loc[row,'ttmto_homol']=='TORNILLO DE EXPANSION')
        |(PPE.loc[row,'ttmto_homol']=='PLACA DE LEVANTE')
        |(PPE.loc[row,'ttmto_homol']=='PROFILAXIS PERIODONTAL')
        |(PPE.loc[row,'ttmto_homol']=='REPARACION TUBO ESTANDAR')
        |(PPE.loc[row,'ttmto_homol']=='PROTESIS TEMP')
        |(PPE.loc[row,'ttmto_homol']=='PROTESIS TEMPORAL  4/8 DIENTES')
        |(PPE.loc[row,'ttmto_homol']=='PROTESIS TEMPORAL 1')
        |(PPE.loc[row,'ttmto_homol']=='RETIRO BRACKETS ARCADA SUPERIOR')
        |(PPE.loc[row,'ttmto_homol']=='RETIRO BRACKETS ARCADA INFERIOR')
        |(PPE.loc[row,'ttmto_homol']=='FLUOR')
        |(PPE.loc[row,'ttmto_homol']=='REPARACION AUTOLIGADO METALICO') 
        |(PPE.loc[row,'ttmto_homol']=='CAMBIO DE NILON (POR UNIDAD DENTAL) SOBREDENTADURA CON BARRA(INCLUY')
        |(PPE.loc[row,'ttmto_homol']=='ATACHES DE PRECISION')
        ):
        PPE.loc[row,'Semana'] = 1

    # TEST "SEM 2 Y TRATAMIENTOS QX":
    elif((PPE.loc[row,'ttmto_homol']=='NUCLEOS')
        |(PPE.loc[row,'ttmto_homol']=='PROVISIONALES')
        |(PPE.loc[row,'ttmto_homol']=='REMOCIÓN PROTESIS FIJA P')
        |(PPE.loc[row,'ttmto_homol']=='INJERTO OSEO GRANULADO 05 GRMS')
        |(PPE.loc[row,'ttmto_homol']=='MEMBRANA REABSORBIBLE')
        |(PPE.loc[row,'ttmto_homol']=='INJERTO OSEO GRANULADO  05 GRMS')
        |(PPE.loc[row,'ttmto_homol']=='INJERTO OSEO GRANULADO 1 GRAMO')
        |(PPE.loc[row,'ttmto_homol']=='REGENERACION OSEA (SIN HUESO NI')
        |(PPE.loc[row,'ttmto_homol']=='ELEVACION  AL SENO MAXILAR TRAUMATICA (INCLUYE INJERTO OSEO  + MEMBRANA)')
        |(PPE.loc[row,'ttmto_homol']=='EXODONCIA QUIRURGICA DE CORDAL (INCLUIDO SEMIERUPCIONADO)')
        |(PPE.loc[row,'ttmto_homol']=='EXODONCIA CORDAL SIMPLE')
        |(PPE.loc[row,'ttmto_homol']=='GINGIVECTOMIA Y GINGIVOPLASTIA X POR CUADRANTE')
        |(PPE.loc[row,'ttmto_homol']=='FRENILLECTOMIA LINGUAL Y LABIAL')
        |(PPE.loc[row,'ttmto_homol']=='BLANQUEAMIENTO DIENTE NO VITAL (TRATAMIENTO COMPLETO)')
        ):
        PPE.loc[row,'Semana'] =2

    elif (PPE.loc[row,'ttmto_homol']=='VENTANA QUIRURGICA'):
        PPE.loc[row,'Semana']=6
        
    # TEST "RETENEDORES": 
    elif ((PPE.loc[row,'ttmto_homol']=='RETENEDOR ESSIX O ESTETICOS')
    |(PPE.loc[row,'ttmto_homol']=='RETENEDOR FIJO 3X3 EN ALAMBRE')
    |(PPE.loc[row,'ttmto_homol']=='RETENEDOR CONTINUO')
    ):  
        if PPE[(PPE['Presupuesto']==temp_ppto)&(
        (PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA ZAFIRO')
        |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA METALICA')
        |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA ESTETICOS O CERAMICOS')
        |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA AUTOLIGABLES')
        |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA AUTOLIGADOS')
        |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA 4X2')
        )].count()[0]:
            temp = PPE[(PPE['Presupuesto']==temp_ppto)&(
            (PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA ZAFIRO')
            |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA METALICA')
            |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA ESTETICOS O CERAMICOS')
            |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA AUTOLIGABLES')
            |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA AUTOLIGADOS')
            |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA 4X2')
            )]
            temp.reset_index(inplace=True)
            PPE.loc[row,'Semana']=temp.loc[0].at['Cantidad Presupuestada']+temp.loc[0].at['Semana']
        else:
            PPE.loc[row,'Semana']=1
        
    # TEST "MENS ORTO 30 DIAS":
    elif ((PPE.loc[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA METALICA')
    |(PPE.loc[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA ZAFIRO')
    |(PPE.loc[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA ESTETICOS O CERAMICOS')
    |(PPE.loc[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA 4X2')
    ):              
        PPE.loc[row,'Semana']=6
        
    # TEST "MENS ORTO 45 DIAS":
    elif ((PPE.loc[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA AUTOLIGABLES')
    |(PPE.loc[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA AUTOLIGADOS')
    ):
        PPE.loc[row,'Semana']=8
        
    # TEST "MONTAJES DE ORTODONCIA Y APARATOLOGIA":
    elif ((PPE.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA SUPERIOR AUTOLIGABLES')
    |(PPE.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA SUPERIOR ZAFIRO')
    |(PPE.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA INFERIOR ZAFIRO')
    |(PPE.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA INFERIOR AUTOLIGABLES')
    |(PPE.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA INFERIOR AUTOLIGADOS')
    |(PPE.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA SUPERIOR AUTOLIGADOS')
    |(PPE.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA INFERIOR ESTETICOS O CERAMICOS') 
    |(PPE.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA SUPERIOR ESTETICOS O CERAMICOS')
    |(PPE.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA INFERIOR METALICOS')
    |(PPE.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA SUPERIOR METALICOS')
    |(PPE.loc[row,'ttmto_homol']=='QUAD HELIX CON BANDAS')
    ):
        PPE.loc[row,'Semana']=2        
        
    # TEST "BLANQUEAMIENTO":
    elif ((PPE.loc[row,'ttmto_homol']=='BLANQUEAMIENTO DENTAL 1')
    |(PPE.loc[row,'ttmto_homol']=='BLANQUEAMIENTO DENTAL 2')):
        PPE.loc[row,'Semana']=1

    # TEST "RASPAJE CON ALISADO RADICULAR":
    elif ((PPE.loc[row,'ttmto_homol']=='RASPAJE CON ALISADO RADICULAR A CAMPO ABIERTO X CUADRANTE')):
        PPE.loc[row,'Semana']=1
    
    # TEST "IMPLANTE y ELEVACION AL SENO ATRAUMATICA":
    elif ((PPE.loc[row,'ttmto_homol']=='IMPLANTE')|(PPE.loc[row,'ttmto_homol']=='ELEVACION AL SENO ATRAUMATICA (POR CUADRANTE)')):
        if PPE[(PPE['Presupuesto']==temp_ppto)&(PPE['ttmto_homol']=='REGENERACION OSEA (SIN HUESO NI')].count()[0]:
            temp = PPE[(PPE['Presupuesto']==temp_ppto)&(PPE['ttmto_homol']=='REGENERACION OSEA (SIN HUESO NI')]
            temp.reset_index(inplace=True)
            PPE.loc[row,'Semana']=temp.loc[0].at['Semana']+24
        else:
            PPE.loc[row,'Semana']=2
    
    # TEST "REHAB SOBRE IMPLANTES":
    elif ((PPE.loc[row,'ttmto_homol']=='SOBREDENTADURA SOBRE 3 IMPLANTES')
    |(PPE.loc[row,'ttmto_homol']=='SOBREDENTADURA SOBRE 4 IMPLANTES')
    |(PPE.loc[row,'ttmto_homol']=='HIBRIDA SOBRE 6 IMPLANTES')
    |(PPE.loc[row,'ttmto_homol']=='CORONA SOBRE IMPLANTE')
    |(PPE.loc[row,'ttmto_homol']=='CORONA SOBRE IMPLANTE ZIRCONIO')
    |(PPE.loc[row,'ttmto_homol']=='PROVISIONAL ESTETICO ATORNILLADO CON PERFIL DE')
    ):
        if PPE[(PPE['Presupuesto']==temp_ppto)&(PPE['ttmto_homol']=='IMPLANTE')].count()[0]:
            temp = PPE[(PPE['Presupuesto']==temp_ppto)&(PPE['ttmto_homol']=='IMPLANTE')]
            temp.reset_index(inplace=True)
            PPE.loc[row,'Semana']=temp.loc[0].at['Semana']+16
        else:
            PPE.loc[row,'Semana']=1
    
    # TEST "REHAB SOBRE DIENTES"
    elif ((PPE.loc[row,'ttmto_homol']=='CARILLA ESTETICA EN RESINA')
    |(PPE.loc[row,'ttmto_homol']=='CORONA METAL CERAMICA')
    |(PPE.loc[row,'ttmto_homol']=='INCRUSTACION EN CEROMERO')
    |(PPE.loc[row,'ttmto_homol']=='CORONA CON COLLAR')
    |(PPE.loc[row,'ttmto_homol']=='CORONAS EN ZIRCONIO SISTEMA CEREC')
    |(PPE.loc[row,'ttmto_homol']=='CORONA ESTETICAS ALUMINA Y CEREC Y ZIRCONIO')
    ):
        if PPE[(PPE['Presupuesto']==temp_ppto)&(PPE['ttmto_homol']=='BLANQUEAMIENTO DENTAL 1')].count()[0]:
            temp = PPE[(PPE['Presupuesto']==temp_ppto)&(PPE['ttmto_homol']=='BLANQUEAMIENTO DENTAL 1')]
            temp.reset_index(inplace=True)
            PPE.loc[row,'Semana']=temp.loc[0].at['Semana']+3
        else:
            PPE.loc[row,'Semana']=2

    # TEST "PLACA NEUROMIORELAJANTE y PLACA DE BRUXISMO":
    elif ((PPE.loc[row,'ttmto_homol']=='PLACA NEUROMIORELAJANTE ACTIVADA')|(PPE.loc[row,'ttmto_homol']=='PLACA DE BRUXISMO')):
        if PPE[(PPE['Presupuesto']==temp_ppto)&(PPE['ttmto_homol']=='CORONA SOBRE IMPLANTE ZIRCONIO')].count()[0]:
            temp = PPE[(PPE['Presupuesto']==temp_ppto)&(PPE['ttmto_homol']=='CORONA SOBRE IMPLANTE ZIRCONIO')]
            temp.reset_index(inplace=True)
            PPE.loc[row,'Semana']=temp.loc[0].at['Semana']+4
        if PPE[(PPE['Presupuesto']==temp_ppto)&(PPE['ttmto_homol']=='CORONA SOBRE IMPLANTE')].count()[0]:
            temp = PPE[(PPE['Presupuesto']==temp_ppto)&(PPE['ttmto_homol']=='CORONA SOBRE IMPLANTE')]
            temp.reset_index(inplace=True)
            PPE.loc[row,'Semana']=temp.loc[0].at['Semana']+4
        if PPE[(PPE['Presupuesto']==temp_ppto)&(PPE['ttmto_homol']=='CARILLA ESTETICA EN RESINA')].count()[0]:
            temp = PPE[(PPE['Presupuesto']==temp_ppto)&(PPE['ttmto_homol']=='CARILLA ESTETICA EN RESINA')]
            temp.reset_index(inplace=True)
            PPE.loc[row,'Semana']=temp.loc[0].at['Semana']+4
        else:
            PPE.loc[row,'Semana']=1

    # TEST "PROTESIS REMOVIBLES":
    elif ((PPE.loc[row,'ttmto_homol']=='PROTESIS METALICA REMOVIBLE')
    |(PPE.loc[row,'ttmto_homol']=='PROTESIS TOTAL SUPERIOR O INFERIOR')):
        if PPE[(PPE['Presupuesto']==temp_ppto)
        &((PPE['ttmto_homol']=='EXODONCIA DE RESTOS RADICULARES')
        |(PPE['ttmto_homol']=='EXODONCIA RESTO RADICULAR QUIRURGICO'))].count()[0]:
            temp = PPE[(PPE['Presupuesto']==temp_ppto)&((PPE['ttmto_homol']=='EXODONCIA DE RESTOS RADICULARES')
            |(PPE['ttmto_homol']=='EXODONCIA RESTO RADICULAR QUIRURGICO'))]
            temp.reset_index(inplace=True)
            PPE.loc[row,'Semana']=temp.loc[0].at['Semana']+6
        else:
            PPE.loc[row,'Semana']=1
            
    # TEST "INJERTO GINGIVAL LIBRE":
    elif ((PPE.loc[row,'ttmto_homol']=='INJERTO GINGIVAL LIBRE')):
        PPE.loc[row,'Semana']=1
        
    # TEST GINGIVECTOMIA, GINGIVOPLASTIA, ALARGAMIENTO CORONAL (despues de la orto y antes de las carillas)
    elif ((PPE.loc[row,'ttmto_homol']=='GINGIVECTOMIA PARA DISEÑO')
    |(PPE.loc[row,'ttmto_homol']=='ALARGAMIENTO CORONAL (')):
        if PPE[(PPE['Presupuesto']==temp_ppto)&(
        (PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA ZAFIRO')
        |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA METALICA')
        |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA ESTETICOS O CERAMICOS')
        |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA AUTOLIGABLES')
        |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA AUTOLIGADOS')
        |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA 4X2')
        )].count()[0]:
            temp = PPE[(PPE['Presupuesto']==temp_ppto)&(
            (PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA ZAFIRO')
            |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA METALICA')
            |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA ESTETICOS O CERAMICOS')
            |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA AUTOLIGABLES')
            |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA AUTOLIGADOS')
            |(PPE['ttmto_homol']=='MENSUALIDAD ORTODONCIA 4X2')
            )]
            temp.reset_index(inplace=True)
            PPE.loc[row,'Semana']=temp.loc[0].at['Cantidad Presupuestada']+temp.loc[0].at['Semana']
        else:
            PPE.loc[row,'Semana']=1
        
revisiones = PPE[PPE['Semana'] == 0]
print('Presupuestos no-ordenados: ',revisiones['Nº Historia'].unique())
revisiones[['Nº Historia','ttmto_homol']]
```
