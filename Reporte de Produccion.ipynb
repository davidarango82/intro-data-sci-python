```python
import numpy as np
import pandas as pd

pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
pd.set_option('display.max_colwidth', -1)

archivo_PPE = 'Produccion_pendiente_exigible_con_tratamientos_Dentix_Calle_90.csv'
archivo_crono = 'crono_Dentix_Calle_90_1.csv'
precios_df =  pd.read_csv('LISTADO DE SERVICIOS.csv', encoding = 'unicode_escape', sep=',')
```

```python
crono_df = pd.read_csv(archivo_crono, encoding = 'unicode_escape', sep=',')
PPE = pd.read_csv(archivo_PPE, encoding = 'unicode_escape', sep=',')

#Eliminación de Puntos y Revisiones
PPE.set_index('Servicio', inplace=True)
PPE.drop(index='Puntos y Revisiones.  ',inplace=True)
PPE.reset_index(inplace=True)
```

# Cruce Cronogramas - PPE para adjuntar datos actualizados

```python
reporte = pd.merge(PPE,crono_df,how='left',left_on=['Fecha Pres.','Nº Historia','Presupuesto','Servicio'], right_on=['Fecha Pres.','Nº Historia','Presupuesto','Servicio'])
reporte
```

# Homologaciones

```python
reporte['Cantidad Producida'] = reporte['Producidos']/reporte['Importe Pres.']
reporte['Cantidad Pendiente'] = reporte['Cantidad Presupuestada']-reporte['Cantidad Producida']
reporte
```
