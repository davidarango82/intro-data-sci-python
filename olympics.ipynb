```python
import pandas as pd
import numpy as np

df1 = pd.read_csv("olympics.csv", index_col=0, skiprows=1)
#df1.rename(index=str, columns={'№ Summer': '#Summer', '01 !': 'Gold'}))
df1.head(3)
#df1.index #returns the index values (countries)
```

## Rename Columns

```python
df1 = df1.rename(columns={"№ Summer":"# Summer", "01 !":"Gold", "02 !":"Silver", "03 !":"Bronze",
                    "№ Winter":"# Winter", "01 !.1":"Gold.1", "02 !.1":"Silver.1", "03 !.1":"Bronze.1",
                    "№ Games":"# Games", "01 !.2":"Gold.2", "02 !.2":"Silver.2", "03 !.2":"Bronze.2"})
df1.head(3)
```

```python
#Boolean mask:
(df1['Gold']>0).head()
```

```python
#apply boolean mask:
only_gold = df1.where(df1['Gold']>0)
only_gold.head()
```

```python
# counts only cells with values that are not nan: (most functions ignore NaN values)
only_gold['Gold'].count()
```

```python
# if we would like to drop the NaN rows in a series, use the dropna() method:
only_gold = only_gold.dropna()
only_gold.head(3)
```

```python
# there is also a shortcut. if we want to apply a boolean mask and remove the NaN rows, do the following:
only_gold = df1[df1['Gold']>0]
only_gold.head(3)
```

```python
#constructing a boolean mask from two boolean masks and bitwise operators (&, |):
((df1['Gold']>0) | (df1['Gold.1']>0)).head(3)
```

```python
#applying a constructed boolean mask to the dataframe and counting the rows:
len(df1[(df1['Gold']>0) | (df1['Gold.1']==0)])

```

```python
#ex query: names of countries with at least 1 winter gold and 0 summer gold medals:
df1[(df1['Gold']==0) & (df1['Gold.1']>0)]
```
