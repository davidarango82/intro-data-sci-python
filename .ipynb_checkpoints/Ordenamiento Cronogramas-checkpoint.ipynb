```python
# el archivo a ordenar no es crono, es pptos_faltantes!

import numpy as np
import pandas as pd

pd.options.display.max_columns = 999
# pd.options.display.max_rows = 999
pd.set_option('display.max_colwidth', -1)

archivo = 'pptos_faltantes.csv'
crono_df = pd.read_csv(archivo, encoding = 'unicode_escape', sep=',')

##### Presupuestos que hay en el archivo #####
pptos = crono_df['Presupuesto'].unique()
```

```python
crono_df.head(1)
```

```python
#crono_df['Semana']
ppto = 'NÂº 4390 - LUZ ANGELA ECHEVERY'
#crono_df.loc[0].at[crono_df['Presupuesto']==ppto][crono_df['ttmto_homol']=='CARILLA ESTETICA EN RESINA']
temp = crono_df[(crono_df['Presupuesto']==ppto)&(crono_df['ttmto_homol']=='BLANQUEAMIENTO DENTAL 1')]
temp.reset_index(inplace=True)
temp.loc[0].at['Semana']
```

```python
rows=crono_df.shape[0]

for row in range(rows):
    
    temp_ppto=crono_df.loc[row,'Presupuesto']

    # TEST "SEM 1": 
    if ((crono_df.loc[row,'ttmto_homol']=='PROFILAXIS')|(crono_df.loc[row,'ttmto_homol']=='DETARTRAJE')
        |(crono_df.loc[row,'ttmto_homol']=='RESINA FOTOPOLIMERIZADO 1 SUP POSTERIOR Y ANTERIOR')
        |(crono_df.loc[row,'ttmto_homol']=='RESINA FOTOPOLIMERIZADO 2 SUP POSTERIOR Y POSTERIOR')
        |(crono_df.loc[row,'ttmto_homol']=='RESINA FOTOPOLIMERIZADO 3 SUP POSTERIOR O POSTERIOR')
        |(crono_df.loc[row,'ttmto_homol']=='TAC 1 Arcada')|(crono_df.loc[row,'ttmto_homol']=='TAC 2 Arcadas')
        |(crono_df.loc[row,'ttmto_homol']=='DESOPTURACION PARA NUCLEO')|(crono_df.loc[row,'ttmto_homol']=='RETRATAMIENTO ENDODONCIA BIRADICULAR')
        |(crono_df.loc[row,'ttmto_homol']=='TRATAMIENTO DE CONDUCTO BIRADICULAR')|(crono_df.loc[row,'ttmto_homol']=='RETRATAMIENTO ENDODONCIA UNIRADICULAR')
        |(crono_df.loc[row,'ttmto_homol']=='REPARACION DE PT(LABORATORIO)')
        |(crono_df.loc[row,'ttmto_homol']=='CITA TALLADO SELECTIVO X PIEZA')|(crono_df.loc[row,'ttmto_homol']=='REPARACION ESTANDAR (REPARACION METALICOS)')
        |(crono_df.loc[row,'ttmto_homol']=='ESTUDIO CEFALOMETRICO')|(crono_df.loc[row,'ttmto_homol']=='EXODONCIA DE RESTOS RADICULARES')
        |(crono_df.loc[row,'ttmto_homol']=='EXODONCIA SIMPLE COMPLEJIDAD 1-2-3')|(crono_df.loc[row,'ttmto_homol']=='EXODONCIA RESTO RADICULAR QUIRURGICO')
        |(crono_df.loc[row,'ttmto_homol']=='VENTANA QUIRURGICA')|(crono_df.loc[row,'ttmto_homol']=='TORNILLO DE EXPANSION')
        |(crono_df.loc[row,'ttmto_homol']=='PLACA DE LEVANTE')|(crono_df.loc[row,'ttmto_homol']=='PROFILAXIS PERIODONTAL')):
            crono_df.loc[row,'Semana'] = 1

    # TEST "SEM 2": |(crono_df.loc[row,'ttmto_homol']=='')
    elif((crono_df.loc[row,'ttmto_homol']=='NUCLEOS')|(crono_df.loc[row,'ttmto_homol']=='PROVISIONALES')
        |(crono_df.loc[row,'ttmto_homol']=='INJERTO OSEO GRANULADO 05 GRMS')|(crono_df.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA INFERIOR ZAFIRO')
        |(crono_df.loc[row,'ttmto_homol']=='ENTRADA BRACKETS ARCADA SUPERIOR ZAFIRO')|(crono_df.loc[row,'ttmto_homol']=='MEMBRANA REABSORBIBLE')
        |(crono_df.loc[row,'ttmto_homol']=='TRATAMIENTO DE CONDUCTO MULTIRADICULAR')|(crono_df.loc[row,'ttmto_homol']=='INJERTO OSEO GRANULADO  05 GRMS')):
        crono_df.loc[row,'Semana'] =2

    # TEST "MENS ORTO 30 DIAS":
    elif ((crono_df.loc[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA METALICA')|(crono_df.loc[row,'ttmto_homol']=='MENSUALIDAD ORTODONCIA ZAFIRO')):
        crono_df.loc[row,'Semana']=6
        
    # TEST "BLANQUEAMIENTO":
    elif ((crono_df.loc[row,'ttmto_homol']=='BLANQUEAMIENTO DENTAL 1')|(crono_df.loc[row,'ttmto_homol']=='BLANQUEAMIENTO DENTAL 2')):
        crono_df.loc[row,'Semana']=1

    # TEST "RASPAJE CON ALISADO RADICULAR":
    elif ((crono_df.loc[row,'ttmto_homol']=='RASPAJE CON ALISADO RADICULAR A CAMPO ABIERTO X CUADRANTE')):
        crono_df.loc[row,'Semana']=1
    
    # TEST "IMPLANTE":
    elif ((crono_df.loc[row,'ttmto_homol']=='IMPLANTE')):
        if crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='REGENERACION OSEA (SIN HUESO NI')].count()[0]:
            temp = crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='REGENERACION OSEA (SIN HUESO NI')]
            temp.reset_index(inplace=True)
            crono_df.loc[row,'Semana']=temp.loc[0].at['Semana']+24
        else:
            crono_df.loc[row,'Semana']=2
    
    # TEST "REHAB SOBRE IMPLANTES":
    elif ((crono_df.loc[row,'ttmto_homol']=='SOBREDENTADURA SOBRE 3 IMPLANTES')|(crono_df.loc[row,'ttmto_homol']=='SOBREDENTADURA SOBRE 4 IMPLANTES')
         |(crono_df.loc[row,'ttmto_homol']=='CORONA SOBRE IMPLANTE')|(crono_df.loc[row,'ttmto_homol']=='CORONA SOBRE IMPLANTE ZIRCONIO')):
        if crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='IMPLANTE')].count()[0]:
            temp = crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='IMPLANTE')]
            temp.reset_index(inplace=True)
            crono_df.loc[row,'Semana']=temp.loc[0].at['Semana']+16
        else:
            crono_df.loc[row,'Semana']=1
            
    # TEST "PROVISIONAL CON PERFIL DE EMERGENCIA":
    elif ((crono_df.loc[row,'ttmto_homol']=='PROVISIONAL ESTETICO ATORNILLADO CON PERFIL DE')):
        if crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='IMPLANTE')].count()[0]:
            temp = crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='IMPLANTE')]
            temp.reset_index(inplace=True)
            crono_df.loc[row,'Semana']=temp.loc[0].at['Semana']+16
        else:
            crono_df.loc[row,'Semana']=1
    
    # TEST "REHAB SOBRE DIENTES"  
    elif ((crono_df.loc[row,'ttmto_homol']=='CARILLA ESTETICA EN RESINA')|(crono_df.loc[row,'ttmto_homol']=='CORONA METAL CERAMICA')):
        if crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='BLANQUEAMIENTO DENTAL 1')].count()[0]:
            temp = crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='BLANQUEAMIENTO DENTAL 1')]
            temp.reset_index(inplace=True)
            crono_df.loc[row,'Semana']=temp.loc[0].at['Semana']+3

    # TEST "PLACA NEUROMIORELAJANTE":
    elif ((crono_df.loc[row,'ttmto_homol']=='PLACA NEUROMIORELAJANTE ACTIVADA')):
        if crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='CORONA SOBRE IMPLANTE ZIRCONIO')].count()[0]:
            temp = crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='CORONA SOBRE IMPLANTE ZIRCONIO')]
            temp.reset_index(inplace=True)
            crono_df.loc[row,'Semana']=temp.loc[0].at['Semana']+4
        if crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='CARILLA ESTETICA EN RESINA')].count()[0]:
            temp = crono_df[(crono_df['Presupuesto']==temp_ppto)&(crono_df['ttmto_homol']=='CARILLA ESTETICA EN RESINA')]
            temp.reset_index(inplace=True)
            crono_df.loc[row,'Semana']=temp.loc[0].at['Semana']+4


            
crono_df[crono_df['Semana'] == 0]
#crono_df
```
